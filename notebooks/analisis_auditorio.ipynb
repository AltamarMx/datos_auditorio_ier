```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
datos = pd.read_csv('../dat/TT-20190906-20211022.dat',skiprows=[0,2,3],index_col=0,parse_dates=True,na_values='NAN')
del datos['RECORD']
datos
```

```python
nombres = datos.columns
```

```python
def grafica(fecha1,tiempo,columnas,y_lim=(0,50)):
    fecha = parse(fecha1)
    fecha2 = fecha + pd.Timedelta(tiempo)


    fig, ax = plt.subplots(figsize=(12,4))

    ax.set_xlim(fecha,fecha2)
    ax.set_ylim(y_lim)
    ax.set_ylabel('Temperatura [oC]')
    ax.set_xlabel('Tiempo [YY-MM-DD]')

    for columna in columnas:
        ax.plot(datos[columna],label=columna)
    ax.legend()
```

```python
nombres = datos.columns
nombres
```

```python
fecha1 = '2020-11-01'
grafica(fecha1,'7D',nombres)
```

```python
# Ejemplo especificando la fecha, una lista pequenia de columnas y cambiando el rango en y

fecha1 = '2020-11-01'
nombres = ['Tubo_1','VentWN26', 'VentWS27', 'MSin28', 'Acamara29', 'Techoin30', 'Techoout31']
grafica(fecha1,'7D',nombres,y_lim=(10,55))
```
